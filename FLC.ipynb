```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
from matplotlib import pyplot as plt
e = ctrl.Antecedent(np.arange(-1, 1.5, 0.05), 'e')
de = ctrl.Antecedent(np.arange(-1, 1.5, 0.05), 'de')
du = ctrl.Consequent(np.arange(-1, 1.5, 0.05), 'du')
e['NB'] = fuzz.trimf(e.universe, [-1,-1,-0.5])
e['NK'] = fuzz.trimf(e.universe, [-1,-0.5, 0])
e['S'] = fuzz.trimf(e.universe, [-0.5, 0,0.5])
e['PK'] = fuzz.trimf(e.universe, [0, 0.5,1])
e['PB'] = fuzz.trimf(e.universe, [0.5,1,1])

de['NB'] = fuzz.trimf(de.universe, [-1,-1,-0.5])
de['NK'] = fuzz.trimf(de.universe, [-1,-0.5, 0])
de['S'] = fuzz.trimf(de.universe, [-0.5, 0,0.5])
de['PK'] = fuzz.trimf(de.universe, [0, 0.5,1])
de['PB'] = fuzz.trimf(de.universe, [0.5,1,1])

du['NB'] = fuzz.trimf(du.universe, [-1,-1,-0.5])
du['NK'] = fuzz.trimf(du.universe, [-1,-0.5, 0])
du['S'] = fuzz.trimf(du.universe, [-0.5, 0,0.5])
du['PK'] = fuzz.trimf(du.universe, [0, 0.5,1])
du['PB'] = fuzz.trimf(du.universe, [0.5,1,1])
e.view()
de.view()
du.view()
rule1 = ctrl.Rule(e['NB'] & de['NB'], du['NB'])
rule2 = ctrl.Rule(e['NB'] & de['NK'], du['NB'])
rule3 = ctrl.Rule(e['NB'] & de['S'], du['NK'])
rule4 = ctrl.Rule(e['NB'] & de['PK'], du['NK'])
rule5 = ctrl.Rule(e['NB'] & de['PB'], du['S'])

rule6 = ctrl.Rule(e['NK'] & de['NB'], du['NB'])
rule7 = ctrl.Rule(e['NK'] & de['NK'], du['NK'])
rule8 = ctrl.Rule(e['NK'] & de['S'], du['NK'])
rule9 = ctrl.Rule(e['NK'] & de['PK'], du['S'])
rule10 =ctrl.Rule(e['NK'] & de['PB'], du['PK'])

rule11 = ctrl.Rule(e['S'] & de['NB'], du['NK'])
rule12 = ctrl.Rule(e['S'] & de['NK'], du['NK'])
rule13 = ctrl.Rule(e['S'] & de['S'], du['S'])
rule14 = ctrl.Rule(e['S'] & de['PK'], du['PK'])
rule15 = ctrl.Rule(e['S'] & de['PB'], du['PK'])

rule16 = ctrl.Rule(e['PK'] & de['NB'], du['NK'])
rule17 = ctrl.Rule(e['PK'] & de['NK'], du['S'])
rule18 = ctrl.Rule(e['PK'] & de['S'], du['PK'])
rule19 = ctrl.Rule(e['PK'] & de['PK'], du['PK'])
rule20 =ctrl.Rule(e['PK'] & de['PB'], du['PB'])

rule21 = ctrl.Rule(e['PB'] & de['NB'], du['S'])
rule22 = ctrl.Rule(e['PB'] & de['NK'], du['PK'])
rule23 = ctrl.Rule(e['PB'] & de['S'], du['PK'])
rule24 = ctrl.Rule(e['PB'] & de['PK'], du['PB'])
rule25 = ctrl.Rule(e['PB'] & de['PB'], du['PB'])


wm_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4,rule5,rule6,rule7,rule8,
                              rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,
                              rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
wm = ctrl.ControlSystemSimulation(wm_ctrl)
```
